```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
%matplotlib inline

def gaussian_blur(img, kernel_size):
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

kernel_size = 3
image_in = mpimg.imread('Transformation2.jpg')
blur_image_in = gaussian_blur(image_in, kernel_size)
plt.imshow(blur_image_in)
plt.axis('off')
transformation = plt.imshow
plt.savefig('Transformation3.jpg',bbox_inches='tight',transparent=True, pad_inches=0)
```
