```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
%matplotlib inline

def grayscale(img):
    return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

image_in = mpimg.imread('Transformation1.jpg')
gray = grayscale(image_in)
plt.imshow(gray, cmap='gray')
plt.axis('off')
transformation = plt.imshow
plt.savefig('Transformation2.jpg',bbox_inches='tight',transparent=True, pad_inches=0)
```
